```python
with open('ares_out/K2-157_ESPRESSO_2024_aresout.dat', 'r') as file:
    data = file.readlines() 
```

```python
header = data[0].strip().split('\t')
data = data[1:]

rows = [line.strip().split('\t') for line in data]

rows_dict = [dict(zip(header, row)) for row in rows]

wl = [row_dict['WL'] for row_dict in rows_dict]
ep = [row_dict['EP'] for row_dict in rows_dict]
loggf = [row_dict['loggf'] for row_dict in rows_dict]
element = [row_dict['element'] for row_dict in rows_dict]
num = [row_dict['num'] for row_dict in rows_dict]
damp = [row_dict['damp'] for row_dict in rows_dict]
EWsun = [row_dict['EWsun'] for row_dict in rows_dict]
wave = [row_dict['wave'] for row_dict in rows_dict]
depth = [row_dict['depth'] for row_dict in rows_dict]
fwhm = [row_dict['FWHM'] for row_dict in rows_dict]
ew = [row_dict['EW'] for row_dict in rows_dict]
ew_err = [row_dict['EWerr'] for row_dict in rows_dict]
```

```python
print("NUMBER OF WAVELENGTH MATCHED:", len(wl))

from collections import Counter
element_counts = Counter(element)
for element, count in element_counts.items():
    print(f"{element} = {count}")
```

```python
#3000 till 8000 wavelength range
```

```python
#visualize the errors
```

```python
ew_err_list = [float(s) for s in ew_err]
ew_list = [float(s) for s in ew]
err_pro = []

for i in range(len(ew_err_list)):
    err_pr = (ew_err_list[i] / ew_list[i]) * 100
    err_pro.append(err_pr)

#plt.figure(figsize=(10,8))
#plt.scatter(ew_list,ew_err_list)
#plt.scatter(ew_list,err_pro)
print(err_pro)
```

```python
plt.figure(figsize=(10,8))
plt.title("ERRORS OF EWs IN %")
#plt.scatter(ew_list,ew_err_list)
plt.scatter(ew_list,err_pro)
plt.ylabel("EW error in %")
plt.xlabel("EW")
```
